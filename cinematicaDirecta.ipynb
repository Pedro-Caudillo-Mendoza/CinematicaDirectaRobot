# Cinemática directa para robots manipuladores de dos a seis articulaciones.

<img src = 'Ecu_01.png'>

```python
from sympy import *
import numpy as np
init_printing()
#%reset -f
#-----------------------------------------------------------------------------------------------------------------------------
print('INSTRUCCIÓN: USE A, a, d y Q PARA VALORES SIMBÓLICOS DE Alpha, a, d y Theta RESPECTIVAMENTE')
print('___________________________________________________________________________________________')
joins = 1 # Nota importante: Cambiar a Joins
total_joins = int(input('\nINGRESE EL NUMÉRO DE ARTICULACIONES: '))
print('------------------------------------------------------------------------------------------------------------------------------')
while joins <= total_joins:
    print('\nParámetros de la articulación ' + str(joins))
#--------------------------------------------------------------------------------------------------------------------------
    A = input("\nIngrese el parámetro Alpha: ")
    if A == 'A':
        cA, sA = symbols('cos(A) sen(A)')
    else:
        A = float(A)
        cA = round(np.cos((A/180)*np.pi),1)
        sA = round(np.sin((A/180)*np.pi),1)
#--------------------------------------------------------------------------------------------------------------------------
    sa = input("Ingrese el parámetro a: ")
    if sa == 'a':
        sa = symbols('a')
    else:
        sa = float(sa)
#--------------------------------------------------------------------------------------------------------------------------
    sd = input("Ingrese el parámetro d: ")
    if sd == 'd':
        sd = symbols('d')
    else:
        sd = float(sd)
#--------------------------------------------------------------------------------------------------------------------------
    Q = input("Ingrese el parámetro Theta: ")
    if Q == 'Q':
        cQ, sQ = symbols('cos(\Theta) sen(\Theta)')
    else:
        Q = float(Q)
        cQ = round(np.cos((Q/180)*np.pi),1)
        sQ = round(np.sin((Q/180)*np.pi),1)
#--------------------------------------------------------------------------------------------------------------------------
    if joins == 1:
        M1 = Matrix([[  cQ   ,  -sQ  ,   0 ,    sa  ],
                     [ sQ*cA , cQ*cA , -sA , -sA*sd ],
                     [ sQ*sA , cQ*sA ,  cA ,  cA*sd ],
                     [   0   ,   0   ,   0 ,     1  ]])   
        MT0_1 = M1
        print('\nArticulación ' + str(joins) + ' relativa a la base.')
        print('\n'+latex(M1))
        print('-----------------------------------------------------------------------------------------------------------------------------')
#--------------------------------------------------------------------------------------------------------------------------
    if joins == 2:
        M2 = Matrix([[  cQ   ,  -sQ  ,   0 ,    sa  ],
                     [ sQ*cA , cQ*cA , -sA , -sA*sd ],
                     [ sQ*sA , cQ*sA ,  cA ,  cA*sd ],
                     [   0   ,   0   ,   0 ,     1  ]])
        MT0_2 = M1 * M2
        print('\nArticulación ' + str(joins) + ' relariva a la articulación ' + str(joins - 1))
        print('\n'+latex(M2))
        print('\nArticulación ' + str(joins) + ' relativa a la base.')
        print('\n'+latex(MT0_2))
        print('-----------------------------------------------------------------------------------------------------------------------------')
#--------------------------------------------------------------------------------------------------------------------------    
    if joins == 3:
        M3 = Matrix([[  cQ   ,  -sQ  ,   0 ,    sa  ],
                     [ sQ*cA , cQ*cA , -sA , -sA*sd ],
                     [ sQ*sA , cQ*sA ,  cA ,  cA*sd ],
                     [   0   ,   0   ,   0 ,     1  ]])
        MT0_3 = M1*M2*M3
        print('\nArticulación ' + str(joins) + ' relariva a la articulación ' + str(joins - 1))
        print('\n'+latex(M3))
        print('\nArticulación ' + str(joins) + ' relativa a la base.')
        print('\n'+latex(MT0_3))
        print('-----------------------------------------------------------------------------------------------------------------------------')
#--------------------------------------------------------------------------------------------------------------------------
    if joins == 4:
        M4 = Matrix([[  cQ   ,  -sQ  ,   0 ,    sa  ],
                     [ sQ*cA , cQ*cA , -sA , -sA*sd ],
                     [ sQ*sA , cQ*sA ,  cA ,  cA*sd ],
                     [   0   ,   0   ,   0 ,     1  ]])
        MT0_4 = M1*M2*M3*M4
        print('\nArticulación ' + str(joins) + ' relariva a la articulación ' + str(joins - 1))
        print('\n'+latex(M4))
        print('\nArticulación ' + str(joins) + ' relativa a la base.')
        print('\n'+latex(MT0_4))
        print('-----------------------------------------------------------------------------------------------------------------------------')
#--------------------------------------------------------------------------------------------------------------------------  
    if joins == 5:
        M5 = Matrix([[  cQ   ,  -sQ  ,   0 ,    sa  ],
                     [ sQ*cA , cQ*cA , -sA , -sA*sd ],
                     [ sQ*sA , cQ*sA ,  cA ,  cA*sd ],
                     [   0   ,   0   ,   0 ,     1  ]])
        MT0_5 = M1*M2*M3*M4*M5
        print('\nArticulación ' + str(joins) + ' relariva a la articulación ' + str(joins - 1))
        print('\n'+latex(M5))
        print('\nArticulación ' + str(joins) + ' relativa a la base.')
        print('\n'+latex(MT0_5))
        print('-----------------------------------------------------------------------------------------------------------------------------')
#--------------------------------------------------------------------------------------------------------------------------   
    if joins == 6:
        M6 = Matrix([[  cQ   ,  -sQ  ,   0 ,    sa  ],
                     [ sQ*cA , cQ*cA , -sA , -sA*sd ],
                     [ sQ*sA , cQ*sA ,  cA ,  cA*sd ],
                     [   0   ,   0   ,   0 ,     1  ]])
        MT0_6 = M1*M2*M3*M4*M5*M6
        print('\nArticulación ' + str(joins) + ' relariva a la articulación ' + str(joins - 1))
        print('\n'+latex(M6))
        print('\nArticulación ' + str(joins) + ' relativa a la base.')
        print('\n'+latex(MT0_6))
        print('-----------------------------------------------------------------------------------------------------------------------------')
#--------------------------------------------------------------------------------------------------------------------------    
    joins = joins + 1
```
